## Transport-Matrix Physics

RHSMode=3 computes a monoenergetic response matrix. RHSMode=2 integrates over speed to build a $3\times3$ transport matrix relating thermodynamic forces to particle, heat, and flow/current responses. Onsager symmetry and positive entropy production are the standard physics gates for these matrices.
